## 2. Variables and Data Types

Python has several built-in data types. The most common ones you'll encounter in data science are:

- **Strings** (`str`): Text data, enclosed in quotes
- **Integers** (`int`): Whole numbers
- **Floats** (`float`): Decimal numbers
- **Booleans** (`bool`): True or False values

Variables are created by assigning a value with the `=` operator — no need to declare the type explicitly.

```python
# Defining variables of different types
name = "Fathan"        # String
age = 30               # Integer
height = 1.75          # Float
is_student = True      # Boolean

# Check the type of each variable
print(f"name: {name} (type: {type(name).__name__})")
print(f"age: {age} (type: {type(age).__name__})")
print(f"height: {height} (type: {type(height).__name__})")
print(f"is_student: {is_student} (type: {type(is_student).__name__})")
```

Try creating your own variables — for example, your favorite number or your city name.

## 3. Basic Operations

Python supports standard arithmetic operations. These work on both integers and floats.

```python
a = 10
b = 5

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Integer Division:", a // b)
print("Modulo:", a % b)
print("Exponentiation:", a ** b)
```

Experiment with different values of `a` and `b`. What happens when you divide by zero?

## 4. Lists and Dictionaries

**Lists** are ordered collections of items — think of them as arrays. They can hold any type of data and are defined with square brackets.

**Dictionaries** store key-value pairs — useful for representing structured data like a student record. They use curly braces.

```python
# List: an ordered collection
numbers = [1, 2, 3, 4]
print("Numbers:", numbers)
print("First element:", numbers[0])
print("Last element:", numbers[-1])
print("Length:", len(numbers))

# Dictionary: key-value pairs
student = {"name": "Ali", "age": 20, "major": "Computer Science"}
print("\nStudent:", student)
print("Student name:", student["name"])
print("Student age:", student["age"])
```

Try adding a new element to the list with `numbers.append(5)` and a new key to the dictionary with `student["gpa"] = 3.8`.

## 6. Loading a Dataset

Pandas makes it easy to load data from CSV files into a **DataFrame** — a table-like data structure with labeled rows and columns.

> **Note:** You need a file called `data.csv` in your working directory for this to run. You can use any CSV dataset — for example, download one from [Kaggle](https://www.kaggle.com/datasets) or create a simple one with a text editor.

The `.head()` method shows the first 5 rows by default, giving you a quick peek at the data.

```python
# Load a CSV file into a DataFrame
# Replace 'data.csv' with the path to your actual dataset
df = pd.read_csv("data.csv")

# Display the first 5 rows
print(df.head())
```

## 7. Exploring the Dataset

Before doing anything with your data, it's essential to understand its structure. Two key methods:

- `.info()` — shows column names, data types, and non-null counts (helps spot missing values)
- `.describe()` — provides summary statistics for numerical columns (mean, std, min, max, quartiles)

```python
# Data types and missing value overview
print(df.info())
```

```python
# Summary statistics for numerical columns
print(df.describe())
```

Look at the output carefully. Are there any columns with fewer non-null values than the total row count? That indicates missing data — something we'll learn to handle in the next tutorial.

## 8. Basic Visualization

Visualizations help us understand distributions and patterns in the data. A **histogram** shows how values in a column are distributed — where most values fall, whether there are outliers, and whether the distribution is skewed.

```python
# Plot a histogram of a numerical column
# Replace 'age' with a column name from your dataset
df['age'].hist(bins=20, edgecolor='black', figsize=(8, 5))
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.grid(axis='y', alpha=0.5)
plt.show()
```

Try plotting histograms of other numerical columns in your dataset. What patterns do you notice?
